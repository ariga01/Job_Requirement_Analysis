```python
import pandas as pd
from googletrans import Translator
import os
import requests
```

```python
df = pd.read_csv('Output\Early_Data_Cleaned.csv', encoding='utf-8')
```

```python
df
```

```python
df.drop_duplicates(subset=['country','title','company','description'], keep='last', inplace=True)
df.applicants = df.applicants.replace({r"Be among the first 25 applicants": r"25", r".applicants": r"", r"Over.": r""}, regex=True)
df.seniority = df.seniority.replace(to_replace=r"BLANK FIELD", value=r"Full-time", regex=True)
```

```python
df
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df
```

```python
df.dropna(subset='description')
```

```python
for row in range(0,700):
    translator = Translator()
    if df.loc[row, 'country'] == 'Japan':
        df.loc[row, 'title'] = translator.translate(df.loc[row, 'title'], src='ja', dest='en').text
```

```python
df
```

```python
keyword = r'Data.*Analyst|.*Data.*Analyst|Business.*Analyst|.*Business.*Analyst|Business.*Intelligence|.*Business.*Intelligence'
df = df[(df.title.str.match(keyword, case=False))]
df.applicants = df.applicants.astype(int)
```
