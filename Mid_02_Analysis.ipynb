```python
import pandas as pd
```

```python
df = pd.read_csv('Output\Mid_Data_Processed.csv', encoding='utf-8')
```

```python
df
```

```python
for i in range(1,6):
    df.loc[df['description'].str.contains(rf'{i} year|{i} year.|{i} 年|{i} 年.', case=False), f'{i} Year'] = 1
```

```python
skills = [
    'Python', 
    'SQL', 
    'Power BI', 
    'Tableau', 
    'Degree', 
    'Excel', 
    'パイソン', 
    'パワービーアイ', 
    'タブロー', 
    'エクセル'
    ]
for i in skills :
    df.loc[df['description'].str.contains(rf', {i}|,{i}|{i},|{i} ,| {i} | {i}|{i} |{i}|{i}.|.{i}|.{i}.',case=False), f'{i}'] = 1
```

```python
skill = ['GPA','R']
for i in skill :
    df.loc[df['description'].str.contains(rf' {i} ',case=True), f'{i}'] = 1
```

```python
get_sum = ['Python', 'SQL', 'Power BI', 'Tableau', 'Degree', 'Excel', 'R', 'GPA']
for i in get_sum:  
    print(f'{i}:', df[f'{i}'].sum())
```

```python
get_year = ['1 Year', '2 Year', '3 Year', '4 Year', '5 Year']
for i in get_year:  
    print(f'{i}:', df[f'{i}'].sum())
```

```python
df = df.drop(columns=['description','location','employment_type'])
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
df.columns
```

```python
for i in range(0, 1442):
    df.loc[i, 'Python'] = df.loc[i, 'Python'] + df.loc[i, 'パイソン']
    df.loc[i, 'Power BI'] = df.loc[i, 'Power BI'] + df.loc[i, 'パワービーアイ']
    df.loc[i, 'Tableau'] = df.loc[i, 'Tableau'] + df.loc[i, 'タブロー']
    df.loc[i, 'Excel'] = df.loc[i, 'Excel'] + df.loc[i, 'エクセル']
df.drop(columns=['パイソン', 'パワービーアイ', 'タブロー', 'エクセル'], inplace=True)
```

```python
cols = list(df.columns)
a, b = cols.index('GPA'), cols.index('R')
cols[b], cols[a] = cols[a], cols[b]
df = df[cols]
```

```python
df
```

```python
cols = ['Python', 'SQL', 'Power BI', 'Tableau', 'Degree', 'Excel', 'R', 'GPA']
result = []
for col in cols:
    to_list = {
        'skils': col,
        'quantity' : df[f'{col}'].sum()
        }
    result.append(to_list)
```

```python
cols = ['1 Year', '2 Year', '3 Year', '4 Year', '5 Year']
years = []
for col in cols:
    to_list = {
        'years': col,
        'quantity' : df[f'{col}'].sum()
        }
    years.append(to_list)
```

```python
pd.DataFrame(result).sort_values(by='quantity', ascending=False)
```

```python
pd.DataFrame(years).sort_values(by='quantity', ascending=False)
```
